```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time

# Setup driver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# Data lists
names, ratings, strongfoot, weakfoot, position, sidePosition = [], [], [], [], [], []
skill_moves, pace, shooting, passing, dribbling = [], [], [], [], []
defending, physic, popularity, igs, body, age, weight, price = [], [], [], [], [], [], [], []
nation, league, club = [], [], []

try:
    for page in range(1, 10):
        print(f'Page {page}')
        link = f'https://www.futbin.com/players?page={page}&version=all_icons&showStats=Age%2CWeight&gender=men'

        # Try opening the page
        try:
            driver.get(link)
        except Exception as e:
            print(f"Failed to load page {page}: {e}")
            continue

        # Wait for rows
        try:
            WebDriverWait(driver, 15).until(
                EC.presence_of_all_elements_located((By.CLASS_NAME, 'player-row'))
            )
        except:
            print(f"Timeout waiting for players on page {page}")
            continue

        tbody_rows = driver.find_elements(By.CLASS_NAME, 'player-row')

        for row in tbody_rows:
            try:
                names.append(row.find_element(By.CLASS_NAME, 'table-player-name').text)
            except: names.append(None)

            try:
                ratings.append(row.find_element(By.CLASS_NAME, 'player-rating-card-text').text)
            except: ratings.append(None)

            try:
                position.append(row.find_element(By.CLASS_NAME, 'table-pos-main').text)
            except: position.append(None)

            try:
                sidePosition.append(row.find_element(By.CLASS_NAME, 'table-pos').find_element(By.CLASS_NAME, 'text-faded').text)
            except: sidePosition.append(None)

            try:
                price.append(row.find_element(By.CLASS_NAME, 'price').text)
            except: price.append(None)

            try:
                weakfoot.append(row.find_element(By.CLASS_NAME, 'table-weak-foot').text)
            except: weakfoot.append(None)

            try:
                skill_moves.append(row.find_element(By.CLASS_NAME, 'table-skills').text)
            except: skill_moves.append(None)

            try:
                pace.append(row.find_element(By.CLASS_NAME, 'table-key-stats').text)
            except: pace.append(None)

            try:
                shooting.append(row.find_element(By.CLASS_NAME, 'table-shooting').text)
            except: shooting.append(None)

            try:
                passing.append(row.find_element(By.CLASS_NAME, 'table-passing').text)
            except: passing.append(None)

            try:
                dribbling.append(row.find_element(By.CLASS_NAME, 'table-dribbling').text)
            except: dribbling.append(None)

            try:
                defending.append(row.find_element(By.CLASS_NAME, 'table-defending').text)
            except: defending.append(None)

            try:
                physic.append(row.find_element(By.CLASS_NAME, 'table-physicality').text)
            except: physic.append(None)

            try:
                popularity.append(row.find_element(By.CLASS_NAME, 'table-popularity').text)
            except: popularity.append(None)

            try:
                igs.append(row.find_element(By.CLASS_NAME, 'table-in-game-stats').text)
            except: igs.append(None)

            try:
                body.append(row.find_element(By.CLASS_NAME, 'table-height').text)
            except: body.append(None)

            try:
                age.append(row.find_element(By.CLASS_NAME, 'table-age').text)
            except: age.append(None)

            try:
                weight.append(row.find_element(By.CLASS_NAME, 'table-weight').text)
            except: weight.append(None)

            try:
                nation.append(row.find_element(By.CLASS_NAME, 'table-player-nation').find_element(By.TAG_NAME, 'img').get_attribute('title'))
            except: nation.append(None)

            try:
                league.append(row.find_element(By.CLASS_NAME, 'table-player-league').find_element(By.TAG_NAME, 'img').get_attribute('title'))
            except: league.append(None)

            try:
                club.append(row.find_element(By.CLASS_NAME, 'table-player-club').find_element(By.TAG_NAME, 'img').get_attribute('title'))
            except: club.append(None)

            try:
                player_strongfoot = row.find_element(By.CLASS_NAME, 'table-foot').find_element(By.TAG_NAME, 'img').get_attribute('src')
                strongfoot.append('Right' if 'right' in player_strongfoot else 'Left')
            except:
                strongfoot.append(None)

            print(names[-1])  # Print player name to track progress

except Exception as final_err:
    print("Unexpected error:", final_err)

finally:
    driver.quit()
```

```python
icons = pd.DataFrame({
    'Player': names,
    'Rating': ratings,
    'Strong Foot': strongfoot,
    'Position': position,
    'sidePosition' : sidePosition,
    'Price': price,
    'Weak Foot': weakfoot,
    'Skill Moves': skill_moves,
    'Pace': pace,
    'Shooting': shooting,
    'Passing': passing,
    'Dribbling': dribbling,
    'Defending': defending,
    'Physic': physic,
    'Popularity': popularity,
    'IGS': igs,
    'Body': body,
    'Age': age,
    'Weight': weight,
    'Nation': nation,
    'League': league,
    'Club': club
})
icons
```

```python
icons.to_csv('Icons.csv', index=False)
```
